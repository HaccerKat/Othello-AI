```python
import torch
import numpy as np
alpha = 1
a = np.random.dirichlet(np.ones(3) * alpha)
prediction = torch.from_numpy(a)
a = np.random.dirichlet(np.ones(3) * alpha)
target = torch.from_numpy(a)
loss = torch.sum(-target * torch.log(prediction))
loss_control = torch.sum(-target * torch.log(target))
print(prediction)
print(target)
print(loss)
print(loss_control)
```

```python
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
import matplotlib.pyplot as plt

T = np.array([6, 7, 8, 9, 10, 11, 12])
power = np.array([1.53E+03, 5.92E+02, 2.04E+02, 7.24E+01, 2.72E+01, 1.10E+01, 4.70E+00])

xnew = np.linspace(T.min(), T.max(), 300)
# print(xnew)

spl = make_interp_spline(T, power, k=5)
print(spl)
power_smooth = spl(xnew)
# a = np.diag(range(15))
#
# plt.matshow(a)
# plt.show()
plt.plot(xnew, power_smooth)
plt.show()
plt.show()
```
