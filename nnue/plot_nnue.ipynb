```python
import matplotlib.pyplot as plt
import numpy as np
LAYERS = [129, 256, 32, 1]
CNT_LAYERS = len(LAYERS)
WEIGHTS_SZ = 41248
BIASES_SZ = 289
RES_SZ = 418
def plot_weights(weights):
    fig, ax = plt.subplots()
    ax.set_title("Weights")
    ax.set_xlabel("Value")
    ax.set_ylabel("Quantity")
    ax.hist(weights, bins=50, alpha=0.7, color='blue')
    plt.show()

def plot_biases(biases):
    fig, ax = plt.subplots()
    ax.set_title("Biases")
    ax.set_xlabel("Value")
    ax.set_ylabel("Quantity")
    ax.hist(biases, bins=50, alpha=0.7, color='blue')
    plt.show()
```

```python
def main():
    # Example weights, biases, and results
    nnue_name = input("Enter the name of the NNUE model (e.g., '1'): ")
    with open("./models/weights_" + nnue_name + ".txt") as f:
        weights = []
        for i in range(CNT_LAYERS - 1):
            line = f.readline()
            weights += [float(x) for x in line.split()]
    with open("./models/biases_" + nnue_name + ".txt") as f:
        biases = []
        for i in range(CNT_LAYERS - 1):
            line = f.readline()
            biases += [float(x) for x in line.split()]

    print("./models/weights_" + nnue_name + ".txt")
    print("Range: [" + str(min(weights)) + ",", str(max(weights)) + "]")
    print("Standard deviation: " + str(np.std(weights)))
    print("95% confidence interval: [" + str(np.percentile(weights, 2.5)) + ", " + str(np.percentile(weights, 97.5)) + "]")
    abs_1 = [x for x in weights if abs(x) > 0.5]
    print("Number of absolute values > 1: " + str(len(abs_1)))
    print("Percentage of absolute values > 1: " + str(len(abs_1) / len(weights) * 100) + "%")
    plot_weights(np.array(weights))
    plot_biases(np.array(biases))
if __name__ == "__main__":
    main()
# This script visualizes the weights, biases, and results of a neural network.
```
