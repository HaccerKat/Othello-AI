```python
import matplotlib.pyplot as plt
import numpy as np
def plot_diffs(diffs, quantization_val):
    fig, ax = plt.subplots()
    ax.set_title("Differences " + quantization_val)
    ax.set_xlabel("Value")
    ax.set_ylabel("Quantity")
    ax.hist(diffs, bins=50, alpha=0.7, color='blue')
    plt.show()
```

```python
for quantization_val in ["01", "02", "05", "10", "35"]:
    with open("./tests/quantize_" + quantization_val + ".txt") as f:
        diffs = list(map(float, f.readline().split()))
    plot_diffs(np.array(diffs), quantization_val)
    print("Range:", np.min(diffs), "to", np.max(diffs))
    print("Mean:", np.mean(diffs))
    print("Std Dev:", np.std(diffs))
    print("95% confidence interval:", np.percentile(diffs, 2.5), "to", np.percentile(diffs, 97.5))
```
